```python
# 层次化索引
import numpy as np
import pandas as pd
from numpy import nan as NA
from pandas import DataFrame
from pandas import MultiIndex
from pandas import Series
```

```python
data = Series(np.random.randn(10),
              index=[['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'd', 'd'],
              [1, 2, 3, 1, 2, 3, 1, 2, 2, 3]])
data
```

```python
data.index # labels是levels的对应展开
```

```python
data['b']
```

```python
data['b': 'c'] # 闭区间
```

```python
data.loc[['a', 'd']] # 选取索引a和d
```

```python
data.unstack() # 把第二集行索引变成列，因为我们现在是Series一维数组。
```

```python
data.unstack().stack() # 再叠回来
```

```python
frame['Ohio'] # 直接选取列
```

```python
# 构造多级索引
mi = MultiIndex.from_arrays([['Ohio', 'Ohio', 'Colorado'],
                             ['Green', 'Red', 'Green']], 
                            names=['state', 'color'])
print(mi)
```

```python
frame = DataFrame(np.arange(12).reshape((4, 3)),
                  index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]])
frame.columns = mi # 直接设置MultiIndex对象
frame
```

```python
# 重排分级顺序
```

```python
# 重置数据
frame = DataFrame(np.arange(12).reshape((4, 3)),
                  index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                  columns=[['Ohio', 'Ohio', 'Colorado'],
                           ['Green', 'Red', 'Green']])
frame.index.names = ['key1', 'key2']
frame.columns.names = ['state', 'color']
```

```python
frame.swaplevel('key1', 'key2') # 交换索引顺序
```

```python
frame.sort_index(level=1) # 对第一层索引排序，不再用sortlevel函数。
```

```python
frame.swaplevel(0, 1).sort_index(level=0)
```

```python
# 根据级别汇总统计
frame.sum(level='key2') # 根据key2的索引求和，每一列上对应行索引相同的元素相加求和。
```

```python
# level对应行还是列由axis决定。
# 每一列上对应列相同的元素相加求和。
frame.sum(level='color', axis=1)
```

```python
frame = DataFrame({'a': range(7), 'b': range(7, 0, -1),
                   'c': ['one', 'one', 'one', 'two', 'two', 'two', 'two'],
                   'd': [0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
# 把列转为索引
frame2 = frame.set_index(['c', 'd']) # 默认drop为True
frame2
```

```python
# 转索引的同时保留列
frame2 = frame.set_index(['c', 'd'], drop=False)
frame2
```

```python
frame2 = frame.set_index(['c', 'd']) # 默认drop为True
frame2
frame2.reset_index() # 把索引再变回列，如果之前drop为False，会因为列同名产生冲突。
```
